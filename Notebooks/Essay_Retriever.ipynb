```python
import pandas as pd
```

```python
essays = pd.read_csv("../data/output_csv/essays.csv")
adus = pd.read_csv("../data/output_csv/adus.csv")
annotations = pd.read_csv("../data/output_csv/annotations.csv")



# Add id column just for retriving purposes
essays['id'] = essays['essay_id'].str.strip('essay').astype(int)
annotations['id'] = annotations['essay_id'].str.strip('essay').astype(int)
adus['id'] = adus['essay_id'].str.strip('essay').astype(int)
```

```python
def retrieve_text_adus(id):
    text = essays.iloc[id-1]['text']
    ann = annotations[annotations['id']==id].sort_values("adu_label")
    adu = adus[adus['id']==id].sort_values('start_ind')[['claim_type','sentence']].values
    
    
    
    display('ADU Annotations:')
    display(ann)
    print("Essay Text:\n")
    print(text,"\n")
    print("ADUs:\n")
    for claim_type, sentence in adu:
        print(claim_type,sentence,"\n")
    return text, ann, adu
```

```python
# Retrieve and display essay, adus and links by id

text, ann, adu = retrieve_text_adus(id = 388)
```
